```python
#check correlation between population and Alcohol

correlation= df['Population'].corr(df['Alcohol'])
print('Correlation:', correlation)
```

The correlation between population and alcohol is negative, meaning that countries (or region) with large population tends to have lower per capita alcohol consumption

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x='Population',y='Alcohol', data=df,line_kws={'color':'red'})
plt.title ('Population vs Alcohol Consumption (Negative Correlation)')
plt.xlabel ('Population')
plt.ylabel('Alcohol (liter per capita')
plt.show()
```

!. We examined the relationship between population and Alcohol consumption per capita across different countries
2. The correlation coeffecient is negative, indicating that coubtries with larger population tend to have lower per capita alcohol consumption
3. This trend is also visible in the scatter plot with the downward-stoping regression line showing the negative relationship.
4.The correlation is weak suggesting that while population size infleunce alcohol consumption, other factors may also play a role.
